# Environment that returns two circle shape datasets with the same centre
This environment is to return a dataset, which consists of two circle shape datasets. This is to test the capability of ann to fit the concircle dataset.

We directly use the dataset provide by sklearn here.

```python
from sklearn import datasets

import matplotlib.pyplot as plt
```

```python
n_samples = 1500

X,y = datasets.make_circles(n_samples=n_samples, factor=.5,
                                      noise=.05)

X = X/2 + 0.5
```

```python
print X
print y
```

```python
plt.scatter(X[y[:]==0,0], X[y[:]==0,1], color='r')
plt.scatter(X[y[:]==1,0], X[y[:]==1,1], color='g')
plt.show()
```
