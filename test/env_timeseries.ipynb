## Unit Test: time series environment - time_series_repo.py
To test the response of the environment "EnvTimeSeriesRepo", the code here try to run a single episode to fetch a time series dataset.

To enable the unit test, change the path in line 14 of 'env/time_series_repo/' from 'env/time_series_repo' to '../env/time_series_repo/'

```python
%matplotlib inline

import matplotlib.pyplot as plt
import itertools
import sys

if "../" not in sys.path:
  sys.path.append("../") 
from env.time_series_repo import EnvTimeSeriesfromRepo
```

```python
env = EnvTimeSeriesfromRepo()
```

```python
# get the first state
eposide = [env.reset()]
rewards = []

# get the consequent states
for t in itertools.count():
    state, reward, done, _ = env.step(0)
    rewards.append(reward)
    
    if done:
        break
    else:
        eposide.append(state)
    
#print eposide
```

```python
# plot data
plt.plot(eposide[1:len(eposide)-1])
plt.show()

# plot label
plt.plot(rewards)
plt.show()
```
